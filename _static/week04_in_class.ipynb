## List comprehension

**Ex** write the code to create a list of numbers from 0 to 5 with a step of 0.5

more concise way to build a list: **list comprehension**

```python
mylist=[k/2 for k in range(11)]
print(mylist)
```

```python
mylist=[greeting+', dear '+name for greeting in ['Hi','Hello'] for name in ['Xinyi','Simone','Erin'] ]
print(mylist)
```

```python
mylist=[]
for greeting in ['Hi','Hello']:
    for name in ['Xinyi','Simone','Erin']:
        mylist.append(greeting+', dear '+name)
print(mylist)
```

**ex** build `list` with the squares of all even numbers in $[0,20)$

```python
mylist=[]

for i in range(20):
    if i%2==0:
        mylist.append(i**2)

print(mylist)
```

```python
mylist=[i**2 for i in range(20) if i%2==0]
print(mylist)
```

```python
def cuberoot(n):
    return pow(n,1/3)

mylist=[cuberoot(k) for k in range(-5,6)]
print(mylist)
```


**Ex** plot $y=x^3+2x+3$ in $[0,1]$ using 100 steps . Use list comprehension to build `x` and `y`

```python
n=100 # number of steps

x=[k/n for k in range(n+1)]
y=[k**3+2*k+3 for k in x]

plt.figure(figsize=(10,10))
plt.plot(x,y,'r-.',lw=3)
plt.show()
```

**Ex.** plot a circle with radius 2, center (0,0)

use the parametric eq. for a circle: $x=2\ \cos(t), \quad y=2\ \sin(t)$ for $t$ in $[0,2\pi)$

```python
from math import pi,sin,cos
```

```python
n=1000 # number of steps

t=[2*pi*k/n for k in range(n+1)] # creating n steps in which interval
x=[2*cos(z) for z in t]
y=[2*sin(z) for z in t]
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y)
plt.axis('equal')
plt.show()
```

## subplots

`plt.subplot(r,c,p)` # r: row, c:column, p:position

```python
from math import pi,sin,cos,tan,atan
N=100
x=[2*pi*k/N for k in range(N)]
s=[sin(k) for k in x]
c=[cos(k) for k in x]
```

```python
plt.subplot(2,2,1)
plt.plot(x,s,'k')

plt.subplot(2,2,2)
plt.plot(x,s,'r')

plt.subplot(2,2,3)
plt.plot(x,c,'y')

plt.subplot(2,2,4)
plt.plot(x,c,'m')

plt.show()
```
