## regular expressions

```python
import re # regular expressions
import requests
```

```python
url = "https://scassani.github.io/MTH337-S2021/_downloads/74af82a6b009f7e6692546a8bc09cbda/gulliver.txt" 
verne = requests.get(url).text
```

**Ex**: find all the words starting with `c` ending with `o`

without reg. expressions

```python
verne1=verne
skip=".,:;\"'[]\{\}()\n\r<>-_*?!"

for char in skip:
    verne1=verne1.replace(char," ")
    
vernelist=verne1.split()
vernelist[:20]
```

```python
colist= [k for k in vernelist if (k[0]=='c' and k[-1]=='o')]

colist[:20]
        
```

**Ex**: find all the words starting with 5 charactes

```python
fivelist= [k for k in vernelist if len(k)==5]
print(fivelist[:50])
```

with reg. expressions

* `\w` : any word character (letters, numbers, underscore)
* `\b` : word delimiter

find all 5 characters words

```python
fivelist_re = re.findall(r"\b\w\w\w\w\w\b",verne)
print(fivelist_re[:50])
```

```python
fivelist_re = re.findall(r"\b\w{5}\b",verne) # \w 5 times
print(fivelist_re[:50])
```

find all 5 to 8 characters words

```python
fivelist_re = re.findall(r"\b\w{5,8}\b",verne) # from 5 to 8 times
print(fivelist_re[:50])
```

find all 5 or more characters words

```python
fivelist_re = re.findall(r"\b\w{5,}\b",verne) #  5 or more times
print(fivelist_re[:50])
```

find all 8 or less characters words

```python
fivelist_re = re.findall(r"\b\w{,8}\b",verne) #  up to 8 times
print(fivelist_re[:50])
```

All five characters words starting with `c` ending with `o`

```python
fivelist_re = re.findall(r"\bc\w{3}o\b",verne) # \w 5 times
print(fivelist_re[:50])
```

`[]` to indicate a group of possible matches

5 chars, starting with `t` or `T` ending with `e`

```python
fivelist_re = re.findall(r"\b[Tt]\w{3}e\b",verne) # \w 5 times
print(fivelist_re[:50])
```

`*` zero or more times

`+` one or more times

```python
list_re = re.findall(r"\b[Ii]\w*t\b",verne) # \w 5 times
print(list_re[:50])
```

```python
list_re = re.findall(r"\b[Ii]\w+t\b",verne) # \w 5 times
print(list_re[:50])
```

`\d` is for digits

```python
list_re = re.findall(r"\d",verne) # \w 5 times
print(list_re[:50])
```

**Ex.** Find

* All of the words with 13 or 14 characters. How many are there?
* All of numbers (not single digits)
* All of the words starting with a capital letter. (`[a-r]` is all the lower case letters from `a` to `r`)

```python
l=re.findall(r"\b\w{13,14}\b",verne)
len(l)
```

```python
print(l)
```

```python
l=re.findall(r"\b\d+\b",verne)
print(l)
```

```python
c=re.findall(r"\b[A-Z]\w*\b",verne)
print(c[:20])
```

`.` : means everything except the newline character

find what comes after the word `These` ( up to the newline character)

```python
re.findall(r"These.*",verne)
```

`*` and `+` are greedy

```python
mystr = "Breakfast will be served 7:00-9:30 AM, and coffee will also be available in the lobby 10:00 - 11:30 AM"
```

```python
re.findall(r"\d.+ AM\b",mystr)
```

`+?` is the non-greedy approach

```python
re.findall(r"\d.+? AM\b",mystr)
```

`|` means `or`

```python
re.findall(r"\b(\w*ing|\w*ed)\b",verne) # all words ending in -ing or ending in -ed
```

parenthesis can bu used to select subpattern that you want to return

```python
re.findall(r"(\b\w*(ing|ed)\b)",verne) # all words ending in -ing or ending in -ed
```

`\W` is the opposite of `\w`

```python
re.findall(r"\b\W+\b",verne)
```

`^` is the negation of what is in the bracket

```python
re.findall(r"\b[a-c]\w*\b",verne)
```

```python
re.findall(r"\b[^a-c]\w*\b",verne)
```

```python
'by' in re.findall(r"\b[^a-c]\w*\b",verne)
```
